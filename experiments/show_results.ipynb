```python
import numpy as np
import sys; sys.path.append('../')
import jax
import matplotlib.pyplot as plt
import seaborn as sns
from src.utils import save_pickle, load_pickle
import os
```

```python
all_pkl_file = [a for a in os.listdir("./pickle_obj/") if a[-3:] == "pkl"]
all_pkl = {a[:a.index("2025")-1]: load_pickle(f"./pickle_obj/{a}") for a in all_pkl_file}
#all_pkl_file_tra = [a for a in os.listdir("./tranditional/") if a[-3:] == "pkl"]
#all_pkl_tra = {a[:a.index("2025")-1]: load_pickle(f"./tranditional/{a}") for a in all_pkl_file_tra}
```

```python
algorithms_tra = ["edsm", "rpni", "ktails"]
algorithms_der = ["derivative"]
algorithms = algorithms_tra + algorithms_der

res = {}
dataset_sizes = [10, 25, 50, 100, 200]

            
for ds in dataset_sizes:
    res[ds] = {}
    for a in algorithms:
        res[ds][a] = []
        for max_n_states in [10, 20, 30, 40, 50, 60, 70]:   # Con 50 tarda mucho
            row = []
            for n_transitions in [2, 5, 9, 15]:
                p = f"{ds}_{max_n_states}"
                row.append(float(np.mean(list(all_pkl[p]["test_errors"][a].values()))) if p in all_pkl else np.nan)
        
            res[ds][a].append(row)
res
```

```python
dataset_size = 10

algorithms = [("edsm", (0, 0)), ("rpni", (0, 1)), ("ktails", (1, 0)), ("derivative", (1, 1))]
fig, axes = plt.subplots(nrows=2, ncols=2, figsize=(10, 8), layout='constrained')
vmin = np.array([res[dataset_size][a] for a, _ in algorithms]).min()
vmax = np.array([res[dataset_size][a] for a, _ in algorithms]).max()
for a, p in algorithms:
    im = sns.heatmap(res[dataset_size][a][::-1], cmap="mako", cbar=False, annot=True, fmt=".2f", ax=axes[p], vmin=vmin, vmax=vmax)
    axes[p].set_title(a.upper())
    axes[p].set_xticks(ticks=[i+0.5 for i in range(4)], labels=[2, 5, 9, 15])
    axes[p].set_yticks(ticks=[i+0.5 for i in range(7)], labels=[70, 60, 50, 40, 30, 20, 10])
    axes[p].set_xlabel("Transitions Per State")
    axes[p].set_ylabel("States")

fig.suptitle(f"Dataset Size = {dataset_size}")
plt.show()
```

```python
all_pkl_file = [a for a in os.listdir("./hyperparameters_moving_max_states/") if a[-3:] == "pkl"]
all_pkl = {a[:a.index("2025")-1]: load_pickle(f"./hyperparameters_moving_max_states/{a}") for a in all_pkl_file}
```

```python
dataset_size = 10
res = []
for lr in [0.1, 0.05, 0.01, 0.005, 0.001]:
    row = []
    for batch_s in [10, 20, 40, 60, 80, 100]:
        if batch_s <= dataset_size:
            row.append(np.mean(list(all_pkl[f'{dataset_size}-{lr}-{batch_s}']["test_errors"].values())))
        else:
            row.append(np.nan)

    res.append(row)

print(res)
sns.heatmap(res, cmap="mako", cbar=False, annot=True, fmt=".2f")#, vmin=vmin, vmax=vmax)

plt.xticks(ticks=[i+0.5 for i in range(6)], labels=["10", "20", "40", "60", "80", "100"])
plt.yticks(ticks=[i+0.5 for i in range(5)], labels=["0.1", "0.05", "0.01", "0.005", "0.001"])
plt.xlabel("Batch Size")
plt.ylabel("Learning Rate")
```
